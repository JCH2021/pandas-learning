```python
import pandas as pd
import numpy as np
```

```python
a = pd.DataFrame(
    {'AA':[np.NaN,1,2,3],
     'BB':[1,2,np.NaN,4],
     'CC':[4,np.NaN,5,6],
     'DD':[np.NaN,7,8,9]},
    index=['aa','bb','cc','dd']
)
a
```

```python
b = pd.DataFrame(
    {'AA':[np.NaN,1,2,3],
     'BB':[1,2,np.NaN,4],
     'CC':[4,np.NaN,5,6],
     'EE':[np.NaN,7,8,9]},
    index=['aa','bb','ee','dd']
)
b
```

```python
df = a+b #缺失处不参与计算
df
```

```python
df['AA'].sum()
```

```python
df.mean() #缺失值不考虑
```

```python
df.groupby('CC').mean() #取参数指定列有值的行,并未执行求均值操作
```

```python
dfc = df.copy()
dfc.fillna(0) #缺失值用0.0填充，
```

```python
dfc['EE'].fillna('e') #对指定列填充
```

```python
dfc.fillna(df.mean())
```

```python
dfc #填充并未对其本身做出任何变动
```

```python
dfc.dropna(axis=1) #将有空值的列丢掉
```

```python
dfc['AA'].dropna(axis=0) #将AA列有空值的行丢掉
```

```python
dfc.count()
```

```python
df.replace({'AA':np.NaN},1) #值的替换
```
