将两个相似的 DataFrame 结合

```python
df1 = pd.DataFrame(
    {"A": [1.0, np.nan, 3.0, 5.0, np.nan], 
     "B": [np.nan, 2.0, 3.0, np.nan, 6.0]}
)
df2 = pd.DataFrame(
    {
        "A": [5.0, 2.0, 4.0, np.nan, 3.0, 7.0],
        "B": [np.nan, np.nan, 3.0, 4.0, 6.0, 8.0],
    }
)
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2) 
#这样就将第一个数据和第二个数据结合起来，填补了第一个没有而第二个有的数据
#此外还补充了第二个有而第一个没有的数据
```

对象的对齐

```python
sc = pd.Series(np.random.randn(5), index=["a", "b", "c", "d", "e"])
```

```python
s1 = sc[:4]
s1
```

```python
s2 = sc[-4:]
s2
```

```python
s1.align(s2) #两个数据对齐，那么没出现的则为缺失
#参数2 join='outer' or 'left' or 'right' or 'inner'
```

按照 dtype 选择某一列

```python
df = pd.DataFrame(
    {
        "string": list("abc"),
        "int64": list(range(1, 4)),
        "uint8": np.arange(3, 6).astype("u1"),
        "float64": np.arange(4.0, 7.0),
        "bool1": [True, False, True],
        "bool2": [False, True, False],
        "dates": pd.date_range("now", periods=3),
        "category": pd.Series(list("ABC")).astype("category"),
    }
)
df
```
