read_csv() 方法 (pd.)

（read_excel() 类似）

基本参数：

   filepath_or_buffer:可以是某一文件路径、URL 或是任意有可读属性的对象
   
   sep:str，要是用的分隔符，对于 read_csv() 默认为','，对 read_table() 默认为 \t
   
   delimiter:str，默认为 None。sep的可选参数名称
   
   delim_whitespace:bool，默认为 False，指定是否将空格用作分隔符
  
   header:int or list of ints，默认为'infer',用作列名的行号以及数据的开头
   
   names:array-like，默认为 None，要使用的列名称列表
   
   index_col:int,str,sequence of int/str,or False,optional,默认为 None，用作 DataFrame 的行标签的列，以字符串名称或列索引的形式给出
   
   usecols:list-like or callable，默认为 None，返回列的子集
   
   squeeze:bool，默认为 False，如果解析的数据只包含一列，则返回一个序列
   
   prefix:str，默认为 None，无标题时添加到列编号的前缀
   
   mangle_dupe_cols:bool，默认为 True，重复列将被指定为 'X','X.1'…'X.N'
   
   dtype:Type name or dict of column->type，默认为 None，数据或列的数据类型
   
   engine:{'c','python',pyarrow'}，要使用的语法分析引擎
   
   converters:dict，默认为 None，用于转换某些列中的值的函数
   
   true_values:list，默认为 None，值视为真
   
   false_values:list，默认为 None，值视为假
   
   skipinitialspace:bool，默认为 False，在分隔符后跳过空格
   
   skiprows:list-like or integer，默认为 None，文始要跳过的行号或行数
   
   skipfooter:int，默认为0，文末要跳过的行数
   
   nrows:int，默认为 None，要读取的文件行数
   
   low_memory:bool，默认为 True
   
   memory_map:bool，默认为 False
   
   na_vaalues:scalar,str,list-like,or dict，默认为 None，要识别为 NA/NaN 的其他字符串
   
   keep_default_na:bool，默认为 True，解析数据时是否包含默认的 NaN 值
   
   na_filter:bool，默认为 True，检测丢失的值标记
   
   verbose:bool，默认为 False，指示放置在非数字列中的 NA 值的数量
   
   skip_blank_lines:bool，默认为 True，如果为真则跳过空白行，而不是解析为 NaN

to_json() 方法 (pd.Series./pd.DataFrame.)

可选参数：

   path_or_buf:用于写入输出的路径名或缓冲区
   
   orient:pd.Series 的 index/split/records 或 pd.DataFrame 的 split/records/index/columns/values/table
   
   double_precision:编码浮点值时使用的小数位，默认是10
   
   force_ascii:使字符串编码为 ASCII，默认为 True
   
   default_handler:无法将对象转换为适合JSON格式时调用的处理程序
